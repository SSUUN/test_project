```python
# 필요한 라이브러리 임포트
import numpy as np
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

import tensorflow as tf
from sklearn.metrics import precision_score, recall_score, f1_score
import numpy as np
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score,f1_score,confusion_matrix
import datetime 
import numpy as np
import os
import pandas as pd
tf.random.set_seed(2025)
np.random.seed(2025)
import datetime 
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3' 
import tensorflow as tf
import time

gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus:  # gpu가 있다면, 용량 한도를 5GB로 설정
  tf.config.experimental.set_virtual_device_configuration(gpus[0], 
        [tf.config.experimental.VirtualDeviceConfiguration(memory_limit=2*1324)])

# x , y, 데이터 가져오기 

y1=pd.get_dummies(y)
x_rt,x_te,y_tr,y_te=train_test_split(x,y1,test_size=0.2,shuffle=True,stratify =y1)

# x,y 가져오기 

# F1 Score를 계산하는 커스텀 메트릭스 함수
def f1_macro(y_true, y_pred):
    # 예측값에서 가장 큰 값의 인덱스를 추출 (One-Hot Encoding -> 클래스 번호로 변환)
    y_pred_classes = tf.argmax(y_pred, axis=-1)
    y_true_classes = tf.argmax(y_true, axis=-1)
    
    # F1 score 계산
    f1 = tf.py_function(
        lambda y_true, y_pred: f1_score(y_true, y_pred, average='macro'),
        [y_true_classes, y_pred_classes],
        tf.float32
    )
    
    return f1
model = Sequential()

# 입력층과 첫 번째 은닉층
model.add(Dense(256, input_dim=x.shape[1], activation='relu'))

model.add(Dense(128, activation='relu'))

model.add(Dense(32, activation='relu'))

model.add(Dense(15, activation='relu'))



model.add(Dense(5, activation='softmax'))

# 모델 컴파일
model.compile(optimizer='adam', 
              loss='categorical_crossentropy', 
              metrics=[f1_macro])

            
```

```python
model.fit(x_rt, y_tr, epochs=2, batch_size=1, validation_data=(x_te, y_te),validation_batch_size=1)


test_loss, test_acc = model.evaluate(x_te, y_te,batch_size=1)
test_loss,test_acc
```
